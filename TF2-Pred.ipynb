````python
import os
import random
import time
import numpy as np
os.environ['TF_CPP_MIN_LOG_LEVEL'] = '3' 

import tensorflow as tf
import tensorflow.keras
import tensorflow.keras.applications
tf.__version__
````

````python
X = np.load('X.npy')
Y = np.load('Y.npy').reshape(-1)
A = X[-2818*2:]
B = Y[-2818*2:]
````

````python
def data_generator(X, Y):
    for i in range(len(X)):
        yield X[i:i+1], Y[i:i+1]

def eval_model(modelname):
    M = tf.keras.models.load_model(modelname)
    return M.evaluate(data_generator(A, B), steps=len(A))
````
